```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python

# Generate data on commute times.
size, scale = 100, 10
commutes = pd.Series(np.random.gamma(scale, size=size) ** 1.5)

commutes.plot.hist( bins=10
                   , rwidth=.5,color='green')
plt.title('Commute Times for 1,000 Commuters')
plt.xlabel('Counts')
plt.ylabel('Commute Time')
plt.grid(axis='y', alpha=0.35)
```

```python
x_axis=['English','physcs','chemistry','biology','hindi']
plt.ylim(0,100)
y_axis=[70,59,60,84,55]
a=np.arange(len(x_axis))
plt.bar(a,y_axis,color='red')
plt.xticks(a,x_axis)
plt.title('MY ATTENDANCE CHART ',color='red')
plt.grid(axis='y', alpha=0.35,color='black')
plt.grid(axis='x', alpha=0.35,color='green')
```

```python
labels='English','Hindi','Chemistry','Physics','Biology'
sizes=70,55,60,56,80
colors='red','green','yellow','blue','pink'
plt.pie(sizes,colors=colors,labels=labels)
plt.axis('equal')
plt.title('PIE CHART OF SYLLUBES',color='red')
plt.grid(axis='y', alpha=0.35)
plt.grid(axis='x', alpha=0.35)
plt.show()
```

```python
x=np.arange(0.4,4*np.pi,0.1)
y=np.sin(x)
z=np.cos(x)
plt.plot(x,y,x,z)
plt.title('sin(x) & cos(x) of graph',color='green')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend(['sin(x)','cos(x)'])
plt.show()
```
